````python
import pandas as pd
from io import BytesIO

bucket_name = nb_parm.split('/')[0]
list_path_to_file = nb_parm.split('/')[1:]
path_to_file = "/".join(list_path_to_file)

byte_csv_data = base_tools.read_file_from_datalake(bucket_name, path_to_file)
data = BytesIO(byte_csv_data)
df = pd.read_csv(data)

df
````

````python
# === Use your agent ===
import uuid
import ast
from dotenv import load_dotenv
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types
from adk_fuzzy_match_bot.agent import root_agent as fuzzy_match_bot
import warnings
warnings.filterwarnings("ignore", category=ResourceWarning)

load_dotenv("/home/jovyan/notebooks/.env")


# Create a new session service to store state
session_service_stateful = InMemorySessionService()

# Create a NEW session
APP_NAME = "CBS_matching"
USER_ID = "cbs_user"
SESSION_ID = str(uuid.uuid4())
stateful_session = session_service_stateful.create_session(
    app_name=APP_NAME,
    user_id=USER_ID,
    session_id=SESSION_ID,
)
print("CREATED NEW SESSION:")
print(f"\tSession ID: {SESSION_ID}")

runner = Runner(
    agent=fuzzy_match_bot,
    app_name=APP_NAME,
    session_service=session_service_stateful,
)

column = ['feed_word','matching_word','score','reasoning']
new_df = []

i = 1
for feed_word in df.iloc[:,0]:
    
    new_message = types.Content(
        role="user", parts=[types.Part(text=str(feed_word))]
    )
    
    for event in runner.run(
        user_id=USER_ID,
        session_id=SESSION_ID,
        new_message=new_message,
    ):
        if event.is_final_response():
            if event.content and event.content.parts:
                list_ot_output = [_.strip().replace('\n','').replace('```text', '').replace('```','').replace("\\","") for _ in event.content.parts[0].text.split('|')]
                for _ in list_ot_output:
                    if len(_) != 0:
                        new_df.append(ast.literal_eval(_.replace('json', '').strip()))
    print(f"Finish process: {i}: [{feed_word}]: {list_ot_output}")
    i += 1

    break

final_df = pd.DataFrame(new_df, columns=column)
print("### Finish Call AI Agent")
````

````python
from io import StringIO

s3 = base_tools.register_datalake()

# Convert DataFrame to CSV in memory
csv_buffer = StringIO()
final_df.to_csv(csv_buffer, index=False)
csv_buffer.seek(0)

# Upload to MinIO
try:
    s3.put_object(
        Bucket='datalake',
        Key='export/final_cbs.csv',
        Body=csv_buffer.getvalue(),
        ContentType='text/csv'
    )
    print(f"‚úÖ Uploaded to datalake/export/final_cbs.csv")
except Exception as e:
    print(f"‚ùå Failed to upload: {e}")
````
