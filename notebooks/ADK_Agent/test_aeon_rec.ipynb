```python
# === Use your agent ===
import uuid

from dotenv import load_dotenv
from google.adk.runners import Runner
from google.adk.sessions import InMemorySessionService
from google.genai import types
from adk_aeon_rec.agent import root_agent as aeon_rec_agent
import warnings
warnings.filterwarnings("ignore", category=ResourceWarning)

load_dotenv("/home/jovyan/notebooks/.env")

## Setup ##
# Create a new session service to store state
session_service_stateful = InMemorySessionService()

# Create a NEW session
APP_NAME = "Aeon recommendation"
USER_ID = "Test_Aeon"
SESSION_ID = str(uuid.uuid4())
stateful_session = session_service_stateful.create_session(
    app_name=APP_NAME,
    user_id=USER_ID,
    session_id=SESSION_ID,
)
print("CREATED NEW SESSION:")
print(f"\tSession ID: {SESSION_ID}")

runner = Runner(
    agent=aeon_rec_agent,
    app_name=APP_NAME,
    session_service=session_service_stateful,
)
## End of Setup ##

user = ['P001', 'P002', 'P003', 'P004', 'P005']
PERSONA = base_tools.get_table_info("SELECT * FROM EDP.PERSONA")
TRANSACTION = base_tools.get_table_info("SELECT * FROM EDP.TRANSACTION")

summary = io.StringIO()

for i in user:
    PERSONA = base_tools.get_table_info(f"SELECT * FROM EDP.PERSONA WHERE PERSONA_ID = '{i}'").to_markdown()
    TRANSACTION = base_tools.get_table_info(f"SELECT * FROM EDP.TRANSACTION WHERE AEON_ID = '{i}'").to_markdown()
    

    new_message = types.Content(
        role="user", parts=[types.Part(text=f"""User Demographic Data {PERSONA}
        User Transaction Data {TRANSACTION}
        """)]
    )

    for event in runner.run(
        user_id=USER_ID,
        session_id=SESSION_ID,
        new_message=new_message,
    ):
        if event.is_final_response():
            if event.content and event.content.parts:
                print(f"Final Response: {event.content.parts[0].text}")
                summary.write(event.content.parts[0].text)
                
    summary.write(f"########################## END OF {i} ##########################\n\n")
```

```python
summary.seek(0)

# Upload to MinIO
try:
    s3 = base_tools.register_datalake()
    s3.put_object(
        Bucket='datalake',
        Key=f'export/{nb_parm}.txt',
        Body=summary.getvalue(),
        ContentType='text/csv'
    )
    print(f"✅ Uploaded to minio Successfully")
except Exception as e:
    print(f"❌ Failed to upload: {e}")
```
